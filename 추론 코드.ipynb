```python
# run_beit3_finetuning.py íŒŒì¼ ì‹¤í–‰í•  ë•Œ seedê°€ ìˆê¸´ í•˜ì§€ë§Œ, seed ê³ ì •

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = True

seed_everything(42) # Seed ê³ ì •
```

## model

```python
# BEIT-3 modelì„ fine-tuning í• ë•Œ, jsoníŒŒì¼ì´ í•„ìš”
# ë°ì´í„°ì…‹ì´ ë°”ë€” ë•Œ, ë°”ë€Œì–´ì•¼ ë˜ëŠ” ë¶€ë¶„

def make_mcq_json(df, out_path, is_train=False):
    mcq_data = []
    for _, r in df.iterrows():
        # ID â†’ question_id
        num_part = re.findall(r'\d+', r['ID'])[0]
        qid = f"1{num_part}1"

        # img_path ë¡œë¶€í„° íŒŒì¼ëª…ë§Œ ë½‘ê³ , í™•ì¥ì(.jpg) ì œê±°
        filename = os.path.basename(r['img_path'])  # e.g. "TEST_000.jpg"
        name, _ext  = os.path.splitext(filename)    # ("TEST_000", ".jpg")

        # ì§ˆë¬¸+ë³´ê¸° ê²°í•©
        mcq = (
            r['Question']
            + " Choices:"
            + f" A. {r['A']}"
            + f" B. {r['B']}"
            + f" C. {r['C']}"
            + f" D. {r['D']}"
        )

        entry = {
            "image_id":    name,    # <-- í™•ì¥ìë¥¼ ë—€ ìˆœìˆ˜ íŒŒì¼ëª…
            "question":    mcq,
            "question_id": qid,
        }
        if is_train:
            entry["answer"] = r["answer"]
        mcq_data.append(entry)

    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(mcq_data, f, ensure_ascii=False, indent=2)


# ì„¸ì…˜ì— ì €ì¥í•©ë‹ˆë‹¤.

make_mcq_json(train_df, "./json_file/train.json", is_train=True)
make_mcq_json(val_df,   "./json_file/val.json", is_train=True)
make_mcq_json(test_df,  "./json_file/test.json", is_train=False)
```

```python
import os
import json
import pandas as pd
from collections import Counter

# sample_submission.csv ë¡œë“œ
sub = pd.read_csv('./sample_submission.csv')

# ì¶”ë¡ ì— ì‚¬ìš©í•  checkpoint ê²½ë¡œ
epoch_checkpoints = [
    "./finetuned_model_file/checkpoint-1.pth",
    "./finetuned_model_file/checkpoint-3.pth",
    "./finetuned_model_file/checkpoint-4.pth",
    "./finetuned_model_file/checkpoint-best.pth",
]

# ì¶”ë¡  ê²°ê³¼ ì €ì¥ í´ë”
output_dirs = [
    "./e1/",
    "./e2/",
    "./e3/",
    "./e4/",
]

# ì¶”ë¡  ë£¨í”„
for ckpt, out_dir in zip(epoch_checkpoints, output_dirs):
    os.makedirs(out_dir, exist_ok=True)
    print(f"ğŸ”¹ Running inference for checkpoint: {ckpt}")

    !python ./github_folder/DACON_VQA/BEiT-3/run_beit3_finetuning.py \
        --model beit3_large_patch16_480_vqacustom \
        --input_size 480 \
        --task vqacustom \
        --batch_size 64 \
        --sentencepiece_model ./github_folder/model_file/beit3.spm \
        --finetune {ckpt} \
        --data_path . \
        --output_dir {out_dir} \
        --num_max_bpe_tokens 128 \
        --eval

print("âœ… ëª¨ë“  epoch ì¶”ë¡  ì™„ë£Œ!")
```

```python
# ê° ê²°ê³¼ json ê²½ë¡œ
json_paths = [os.path.join(out_dir, "submit_vqacustom_test.json") for out_dir in output_dirs]

# ì˜ˆì¸¡ ê²°ê³¼ ëª¨ìœ¼ê¸°
all_preds = []
for path in json_paths:
    if not os.path.exists(path):
        raise FileNotFoundError(f"âŒ ê²°ê³¼ íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤: {path}")
    with open(path) as f:
        preds = json.load(f)

    answers = []
    for d in preds:
        ans = d['answer']
        if not ans or ans.lower() not in ['a','b','c','d']:
            ans = 'A'
        else:
            ans = ans.upper()
        answers.append(ans)

    all_preds.append(answers)


# ê°¯ìˆ˜ ì¼ì¹˜ í™•ì¸
num_samples = len(all_preds[0])
for i, preds in enumerate(all_preds):
    if len(preds) != num_samples:
        raise ValueError(f"âŒ {i}ë²ˆì§¸ ì˜ˆì¸¡ ê¸¸ì´ê°€ ë‹¬ë¼ìš”!")
```

```python
# í•˜ë“œë³´íŒ…
final_answers = []
for i in range(num_samples):
    votes = [preds[i] for preds in all_preds]
    most_common = Counter(votes).most_common(1)[0][0]
    final_answers.append(most_common)

# ì œì¶œ íŒŒì¼ ìƒì„±
sub['answer'] = final_answers
sub.to_csv('./beit3_hardvote_sub.csv', index=False)

print("âœ… í•˜ë“œë³´íŒ… ê²°ê³¼ ìƒì„± ì™„ë£Œ: beit3_hardvote_sub.csv")

# ë‹¤ìš´ë¡œë“œ
from google.colab import files
files.download('./beit3_hardvote_sub.csv')
```
